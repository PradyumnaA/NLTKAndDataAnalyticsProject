```python
import re
def cleanResume(txt):
    """
    Create a resume dataset remove all the unwanted special charecters here sub stands for substitute
    """
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
someString="Immameaaaooowwww #Genius #Okau #Cokaywebsite like this http://helloword and access it okaucokayitemshouldnotrepeatba@gmail.com"
cleanResume(someString)

```

```python
dataframe["Resume"]=dataframe["Resume"].apply(lambda x:cleanResume(x))
```

```python
from sklearn.preprocessing import LabelEncoder
#Hey encode the category values as integers and store it
#You make category values as unique and store it as columns
label_encoder = LabelEncoder()
dataframe['Category'] = label_encoder.fit_transform(dataframe['Category'])
print(dataframe['Category'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#Convert the words to english charecters
tfidf=TfidfVectorizer(stop_words='english')
#Create a term frequency inverse document frequency of charecters

tfidf.fit(dataframe["Resume"])
requiredText=tfidf.transform(dataframe["Resume"])
print(requiredText)
```

```python
"""
Working of tfidf vectorizer
Say suppose we have sentence "Data Analytics course "
Matrix form can be given as 

|--
|1 0 0
|0 1 0
|0 0 1
Here first bit represents Data
Next is Analytics 
Another is course
|--
"""
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(requiredText,dataframe['Category'],test_size=0.35,random_state=50)
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

# Create a KNeighborsClassifier with distance-weighted neighbors
clf = OneVsRestClassifier(KNeighborsClassifier(weights='distance'))

# Train the classifier
clf.fit(X_train, Y_train)

# Print the classifier
print(clf)

# Make predictions on the test set
Y_Pred = clf.predict(X_test)
print(accuracy_score(Y_test,Y_Pred))
```
